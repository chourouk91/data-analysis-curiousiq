```python
#Fill empty (nan) values of host_response_time to N/A

```

```python
listning['host_response_time'].fillna('N/A', inplace=True)
listning.head()
```

```python
# Find how many listings per host_response_time in a pie chart (hint: unique() and value_counts())
```

```python
from matplotlib import pyplot
labels = list(listning['host_response_time'].unique())
sizes = list(listning['host_response_time'].value_counts())
explode = (0.1, 0, 0, 0, 0) 

fig1, ax1 = plt.subplots(1,1,figsize=(4,4))
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90 ,colors= ['red', 'yellow', 'pink', 'brown', 'green'])
handles, labels = ax1.axes.get_legend_handles_labels()
ax1.legend(handles, labels, prop={'size':6},
          bbox_to_anchor=(0.5,0.90),
          bbox_transform=fig1.transFigure)
ax1.axis('equal')  

plt.show()
```
